```python
# Let's plot the function f(x) = x from seaborn on the x-axis (0, 1) and y_axis (0, 1)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

x = np.linspace(0, 1, 100)

# Make sure to use nicer seaborn style

sns.set(style="whitegrid")

plt.plot(x, x)

plt.show()
```

```python
# Now plot the functions x^2, sqrt(x) and cube root x on the same plot and include a legend
# Add a title to the plot
plt.title('Various Noise Injection Functions')

# Colour the lines according to a pastel color palette that corresponds to severity of noise injection. x^2 should be lowest, then x, then sqrt(x), then cube root x
plt.plot(x, x**2, label='x^2', color='lightblue')
plt.plot(x, x, label='x', color='blue')
plt.plot(x, np.sqrt(x), label='x^(1/2)', color='darkblue')
plt.plot(x, x**(1/3), label='x^(1/3)', color='black')
plt.legend()

# Save this plot as noise_injection_functions.png with high resolution dpi = 300
plt.savefig('noise_injection_functions.png', dpi=300)
plt.show()
```
