## Plot 3 Layer Results
LSTM, AdaBelief vs Base

```python

# Define the list of optimizers and their corresponding files
functions = ['base', 'x^2', 'x', 'x^(1/2)', 'x^(1/3)']
files = ['data/3layer/lstm_adabelief_log_3layer',
         'data/3layer/lstm_adabelief_lm_log_3layer_xx',
         'data/3layer/lstm_adabelief_lm_log_3layer_x',
         'data/3layer/lstm_adabelief_lm_log_3layer_sqrtx',
         'data/3layer/lstm_adabelief_lm_log_3layer_cuberoot']

# Read and plot data for each function
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # Adjusted for smaller, square plots side by side
legend_labels = ["AdaBelief", "AdaBelief-LM (x^2)", "AdaBelief-LM (x)", "AdaBelief-LM (x^(1/2))", "AdaBelief-LM (x^(1/3))"]

# Training Set Perplexity Plot
for func, file, label in zip(functions, files, legend_labels):
    # Assuming the use of example data for simulation, using io.StringIO for demonstration.
    df_train = pd.read_csv(file)['train_ppl']
    ax1.plot(df_train, label=label, linestyle='--')  # Making lines dashed

ax1.set_title('Training Set Perplexity Per Epoch')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Training Perplexity')
ax1.set_ylim(35, 100)  # Setting y-axis limits as specified
ax1.legend(title="Optimizer Variant", fontsize='x-small', title_fontsize='small', loc='upper right', frameon=True)

# Validation Set Perplexity Plot
for func, file, label in zip(functions, files, legend_labels):
    # Assuming the use of example data for simulation.
    df_valid = pd.read_csv(file)['valid_ppl']
    ax2.plot(df_valid, label=label, linestyle='--')  # Making lines dashed

ax2.set_title('Validation Set Perplexity Per Epoch')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Validation Perplexity')
ax2.set_ylim(50, 90)  # Setting y-axis limits as specified
ax2.legend(title="Optimizer Variant", fontsize='x-small', title_fontsize='small', loc='upper right', frameon=True)

plt.tight_layout()  # Adjust layout to not overlap
plt.show()

fig.savefig('lstm_3layer_loss_curve.png', format='png', dpi=300)  # High-resolution save
```

# Minimum loss charts


## 1 layer


```python
files = ['data/1layer/lstm_adabelief_log_1layer',
         'data/1layer/lstm_adabelief_log_1layer_seed6',
         'data/1layer/lstm_adabelief_log_1layer_seed42',
         'data/1layer/lstm_adabelief_lm_log_1layer_xx',
         'data/1layer/lstm_adabelief_lm_log_1layer_xx_seed6',
         'data/1layer/lstm_adabelief_lm_log_1layer_xx_seed42',
         'data/1layer/lstm_adabelief_lm_log_1layer_x',
         'data/1layer/lstm_adabelief_lm_log_1layer_x_seed6',
         'data/1layer/lstm_adabelief_lm_log_1layer_x_seed42',
         'data/1layer/lstm_adabelief_lm_log_1layer_sqrtx',
         'data/1layer/lstm_adabelief_lm_log_1layer_sqrtx_seed6',
         'data/1layer/lstm_adabelief_lm_log_1layer_sqrtx_seed42',
         'data/1layer/lstm_adabelief_lm_log_1layer_cuberoot',
         'data/1layer/lstm_adabelief_lm_log_1layer_cuberoot_seed6',
         'data/1layer/lstm_adabelief_lm_log_1layer_cuberoot_seed42']

optimizer_names = ["AdaBelief", "AdaBelief-LM (x^2)", "AdaBelief-LM (x)", "AdaBelief-LM (x^(1/2))", "AdaBelief-LM (x^(1/3))"]

data = []

# Read each file and process it
for i, file_path in enumerate(files):
    optimizer_index = i // 3  # Integer division to cycle through optimizer names
    optimizer_name = optimizer_names[optimizer_index % len(optimizer_names)]

    df = pd.read_csv(file_path)
    
    # Extract minimum values
    min_train_ppl = df['train_ppl'].min()
    min_valid_ppl = df['valid_ppl'].min()
    
    # Extract optimizer variant and seed from file_path
    parts = file_path.split('_')
    optimizer = parts[-3]  # This may need adjustment based on actual path structure
    
    # Determine the seed based on the file name
    seed = '141'
    if 'seed6' in file_path:
        seed = '6'
    elif 'seed42' in file_path:
        seed = '42'
    
    # Append the data
    data.append({'Optimizer': optimizer_name, 'Seed': seed, 'Min Train PPL': min_train_ppl, 'Min Valid PPL': min_valid_ppl})

# Convert list to DataFrame
df_data = pd.DataFrame(data)

# Calculating the averages
train_avg = df_data.groupby('Optimizer')['Min Train PPL'].mean().reset_index()
valid_avg = df_data.groupby('Optimizer')['Min Valid PPL'].mean().reset_index()

print(train_avg)
print(valid_avg)

fig, axs = plt.subplots(1, 2, figsize=(14, 7))  # Share y-axis to keep the scale consistent

sns.stripplot(ax=axs[0], data=df_data, x='Optimizer', y='Min Train PPL', hue='Seed', dodge=False, jitter=False, size=8, alpha=0.8)
axs[0].set_title('Minimum Training Perplexity')
axs[0].tick_params(axis='x', rotation=20)
axs[0].set_xlabel('')  # Hiding the x-axis label
axs[0].set_ylabel('Perplexity')


# Plot Minimum Validation Perplexity
sns.stripplot(ax=axs[1], data=df_data, x='Optimizer', y='Min Valid PPL', hue='Seed', dodge=False, jitter=False, size=8, alpha=0.8)
axs[1].set_title('Minimum Validation Perplexity')
axs[1].tick_params(axis='x', rotation=20)
axs[1].set_xlabel('')  # Hiding the x-axis label
axs[1].set_ylabel('Perplexity')



# Remove the legend from the first subplot to have only one legend for the whole figure
axs[0].get_legend().remove()


# Adjust the legend of the first plot to apply for the whole figure
handles, labels = axs[0].get_legend_handles_labels()

plt.tight_layout()
plt.show()

fig.savefig('lstm_1layer_minppl.png', format='png', dpi=300)  # High-resolution save
```
